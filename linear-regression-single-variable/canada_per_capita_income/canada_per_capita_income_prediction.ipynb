```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Capitale Income (US$)")
plt.scatter(df.year,df['per capita income (US$)'],color="red",marker="+")
```

```python
reg = LinearRegression()
reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
predicted_income_2018 = reg.predict([[2018]])
print(f"Predicted Income for 2018: ${predicted_income_2018}")
```

```python
print('y= m x + b')
print('slope (b): ',reg.coef_)
print('intercept (m): ',reg.intercept_)
```

```python
print("y= m * Year + b")
y= reg.coef_ * 2018 + reg.intercept_
print("y= ",y)
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Capitale Income (US$)")
plt.scatter(df.year,df['per capita income (US$)'],color="red",marker="+")
plt.plot(df.year,reg.predict(df[['year']]),color="blue")
```

```python
df.loc[len(df)] = [2018,predicted_income_2018]
```
